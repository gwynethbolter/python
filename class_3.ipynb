lists are mutable (or able to be modified)

```python
fruits[1] = "kiwi"
fruits
```

```python
rainbow = ["red", "orange", "yellow", "green", "light blue", "blue", "violet"]

rainbow.append("purple")
rainbow
```

```python
rainbow.extend(["magenta", "pink"])
# extend adds more than 1 value onto the list. append is 1 value
rainbow
```

```python
# if you try to extend without putting the values in a list
# it will break your string up into chars
rainbow.extend("pink")
rainbow
```

```python
rainbow.insert(6, "indigo")
rainbow
# this inserts, does not replace
```

```python
rainbow.remove("indigo")
rainbow
# remove the FIRST instance of indigo
```

```python
del rainbow[0:3]
rainbow
```

a set stores unique values that are unordered and do not allow duplicates

```python
things = {'coat', 'lock', 'lock', 'apple'}
things
# this will remove duplicates and order them
```

```python
postal_codes = ["M5R", "M5V", "M1M"]
set(postal_codes)
# change a list to a set
```

sets cannot be sliced, since they are unordered

but, sets are mutable so we can add and remove items

```python
things.add('lock')
things
```

```python
things.add('pear')
things
```

```python
things.remove('pear')
things
```

```python
olympic_flag = {"red", "green", "yellow", "blue", "black"}
rainbow = {"red", "orange", "yellow", "green", "blue", "purple"}
rainbow.union(olympic_flag)
```

```python
rainbow.intersection(olympic_flag)
```

```python
rainbow.difference(olympic_flag)
# what makes rainbow different from olympic flag
```

```python
olympic_flag.difference(rainbow)
# what makes olympic flag different from rainbow
```
