what if we want to loop through more than one thing at a time?

```python
stops = ["Sheppard-Yonge", "Bayview", "Bessarion", "Leslie", "Don Mills"]

for idx, stop in enumerate(stops):
    print(f'Stop {idx + 1} is {stop}')
    # enumerate gives us two values, the index and the stop itself
```

```python
prices = {'apple': 2, 'banana': 1, 'orange': 3}

for fruit, price in prices.items():
    print(fruit,price)
```

remember that items provides us with key item pairs

```python
for fruit, price in prices.items():
    if price == 1:
        print(f"holy cheap {fruit}")
    else:
        print(f"{fruit} costs ${price}")
```

nested loops

```python
for i in range(1, 3):
    for j in range(1, 3):
        print(i, "*", j, "=", i*j)
```

```python
counter = 4

while counter > 0:
    print(counter)
    counter -= 1
```

```python
while counter > 0:
    print(counter)
    if counter == 3:
        break
    counter -= 1
```

```python
for i in range(1,6):
    if i % 2 != 0:
        continue 
        # this means skip the rest of the loop
    print(i)
```
