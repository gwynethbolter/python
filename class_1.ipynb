keyword arguments specify variables in the function

```python
def calc_total_bill(price, tax_rate = 0.13, tip_rate = 0.2):
    '''
   calculates total bill
   this is a docstring
    '''
    tax = price * tax_rate
    tip = price * tip_rate
    return price + tax + tip
```

```python
calc_total_bill(100)
```

```python
calc_total_bill(100, tip_rate = 0.22)
```
