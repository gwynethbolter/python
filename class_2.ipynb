```python
def greet_user():
    name = input("enter your name")
    return(f"Hello, {name}!")

greet_user()
```

```python
def user_colour():
    colour = input("enter your favourite colour!")
    return(f"Your favourite colour is {colour}!")

user_colour()
```

```python
def user_name():
    first_name = input("Enter your first name")
    last_name = input("Enter your last name")
    return(f"your name is: {first_name} {last_name}")

user_name()
```

control flow determines what will be executed when, ie an if statement

```python
year = 2026

if year >= 2000:
    print("We are in the 21st century")
elif year >= 2000:
    print("We are in the 20th century")
else:
    print("We are not in the 20th or 21st century")
```

```python
temp = -49
if temp < 0:
    print("It is freezing cold!")
elif temp >= 1 and temp <= 15:
    print("It is cold!")
elif temp >= 16 and temp <= 30:
    print("It is hot!")
elif temp < 45:
    print("It is really hot!")
else:
    print("It is extremely hot!")
```
